# Gr√°fico de Notas de Aluno [Histograma]

```python
# Gerando notas aleat√≥rias
import numpy as np
np.random.randint(42)
notas = np.random.normal( 50, 10, 500 )

# Gr√°fico
plt.figure(figsize=(10,8))
plt.hist( notas, bins=50, edgecolor='navy', color='royalblue' )
plt.axvline( notas.mean(), color='indigo', linestyle='--', linewidth='2')
plt.legend(['M√©dia das Notas'])
plt.title('Distribui√ß√£o de notas dos alunos de turma')
plt.xlabel('Notas')
plt.ylabel('N√∫mero de alunos')
plt.show()
```

# Gr√°fico da Composi√ß√£o de Carteira de Investimento [Pizza]

```python
# Dados
ativos = ['A√ß√µes', 'Renda Fixa', 'Ouro', 'Criptomoedas']
investimentos = [ 30, 10, 15, 25 ]
destaque = [0.1, 0, 0, 0]

plt.figure(figsize=(7, 7))
plt.pie(investimentos, labels=ativos, colors=['skyblue', 'purple', 'gold', 'red'], autopct='%1.1f%%', explode=destaque)
plt.axis('equal')
plt.title("Distribui√ß√£o de Investimentos")
plt.show()
```

# Gr√°fico de rela√ß√£o entre altura e peso [Dispers√£o]

```python
alturas = np.random.normal ( 1.67, 10, 50 )
pesos = ( alturas * 0.5 ) + np.random.normal ( 0, 5, 50 )

plt.figure(figsize=(6,6))
plt.scatter( alturas, pesos, color='orange', s=20, alpha=0.5 )
plt.axvline( alturas.mean(), color='red', linestyle='--', linewidth='2')
plt.axvline( pesos.mean(), color='blue', linestyle=':', linewidth='2')
plt.legend(['Dispers√£o', 'M√©dia das Alturas', 'M√©dia dos Pesos'])
plt.title('Rela√ß√£o Altura x Peso')
plt.xlabel('Altura em cm')
plt.ylabel('Peso em Kg')
plt.show()
```

# Analise - Titanic

Analisar novamente o dataset do titanic e formular um gr√°fico para obter um insight

## Conclus√£o Obtida:

A) O n√∫mero de mulheres que sobreviveram foi significativamente maior do que o n√∫mero de homens que sobreviveram

```python
# Bibliotecas
import pandas as pd

# Importanto DF
df = pd.read_csv('titanic.csv')

# Sobreviventes por sexo

sobrevivencia_sexo = df.groupby('Sex')['Survived'].sum()

labels = {'male': 'Homens', 'female': 'Mulheres'}

# Gr√°fico de barras
plt.bar([labels[i] for i in sobrevivencia_sexo.index], sobrevivencia_sexo.values)
plt.xlabel("Sexo")
plt.ylabel("Sobreviventes")
plt.title("Sobreviventes por Sexo")
plt.show()
```

```python
# Taxa de sobreviv√™ncia 
taxa_sobrevivencia_sexo = df.groupby('Sex')['Survived'].mean().round(2) * 100

plt.pie(taxa_sobrevivencia_sexo.values, labels=[labels[i] for i in taxa_sobrevivencia_sexo.index],autopct='%1.1f%%', colors=['lightblue', 'pink'])
plt.xlabel("Sexo")
plt.ylabel("Taxa")
plt.title("Taxa de sobreviv√™ncia")
plt.show()
```
