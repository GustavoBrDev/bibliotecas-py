# Gráfico de Notas de Aluno [Histograma]

```python
# Gerando notas aleatórias
import numpy as np
np.random.randint(42)
notas = np.random.normal( 50, 10, 500 )

# Gráfico
plt.figure(figsize=(10,8))
plt.hist( notas, bins=50, edgecolor='navy', color='royalblue' )
plt.axvline( notas.mean(), color='indigo', linestyle='--', linewidth='2')
plt.legend(['Média das Notas'])
plt.title('Distribuição de notas dos alunos de turma')
plt.xlabel('Notas')
plt.ylabel('Número de alunos')
plt.show()
```

# Gráfico da Composição de Carteira de Investimento [Pizza]

```python
# Dados
ativos = ['Ações', 'Renda Fixa', 'Ouro', 'Criptomoedas']
investimentos = [ 30, 10, 15, 25 ]
destaque = [0.1, 0, 0, 0]

plt.figure(figsize=(7, 7))
plt.pie(investimentos, labels=ativos, colors=['skyblue', 'purple', 'gold', 'red'], autopct='%1.1f%%', explode=destaque)
plt.axis('equal')
plt.title("Distribuição de Investimentos")
plt.show()
```

# Gráfico de relação entre altura e peso [Dispersão]

```python
alturas = np.random.normal ( 1.67, 10, 50 )
pesos = ( alturas * 0.5 ) + np.random.normal ( 0, 5, 50 )

plt.figure(figsize=(6,6))
plt.scatter( alturas, pesos, color='orange', s=20, alpha=0.5 )
plt.axvline( alturas.mean(), color='red', linestyle='--', linewidth='2')
plt.axvline( pesos.mean(), color='blue', linestyle=':', linewidth='2')
plt.legend(['Dispersão', 'Média das Alturas', 'Média dos Pesos'])
plt.title('Relação Altura x Peso')
plt.xlabel('Altura em cm')
plt.ylabel('Peso em Kg')
plt.show()
```

# Analise - Titanic

Analisar novamente o dataset do titanic e formular um gráfico para obter um insight

## Conclusão Obtida:

A) O número de mulheres que sobreviveram foi significativamente maior do que o número de homens que sobreviveram

```python
# Bibliotecas
import pandas as pd

# Importanto DF
df = pd.read_csv('titanic.csv')

# Sobreviventes por sexo

sobrevivencia_sexo = df.groupby('Sex')['Survived'].sum()

labels = {'male': 'Homens', 'female': 'Mulheres'}

# Gráfico de barras
plt.bar([labels[i] for i in sobrevivencia_sexo.index], sobrevivencia_sexo.values)
plt.xlabel("Sexo")
plt.ylabel("Sobreviventes")
plt.title("Sobreviventes por Sexo")
plt.show()
```

```python
# Taxa de sobrevivência 
taxa_sobrevivencia_sexo = df.groupby('Sex')['Survived'].mean().round(2) * 100

plt.pie(taxa_sobrevivencia_sexo.values, labels=[labels[i] for i in taxa_sobrevivencia_sexo.index],autopct='%1.1f%%', colors=['lightblue', 'pink'])
plt.xlabel("Sexo")
plt.ylabel("Taxa")
plt.title("Taxa de sobrevivência")
plt.show()
```
